```python
def body_mass_index(height, weight):
    return round((weight / height ** 2), 2)

print("Hi, Welcome to BMI calculator.")

while True:
    name = input("Enter your name: ")
    if name.isdigit():
        print("Input invalid. Please enter a string.")
    else:
        break

while True:
    age = input("Enter your Age: ")
    if not age.isdigit():
        print("Input invalid. Please enter a number.")
    else:
        age = int(age)
        if age < 10:
            print("Input invalid. Please enter a valid age.")
        else:
            break

while True:
    try:
        height = float(input("Enter your height (in meters): "))
        if height <= 0:
            print("Height should be a positive number.")
        else:
            break
    except ValueError:
        print("Input invalid. Please enter a number.")

while True:
    try:
        weight = float(input("Enter your weight (in kilograms): "))
        if weight <= 0:
            print("Weight should be a positive number.")
        else:
            break
    except ValueError:
        print("Input invalid. Please enter a number.")

BMI = body_mass_index(height, weight)
BMI = round(BMI, 1)

if BMI < 18.0:
    print(f"Your BMI is: {BMI}. You are underweight.")
    ideal_weight = 18.0 * (height ** 2)
    ideal_weight = round(ideal_weight, 1)
    print(f"You should reach a weight of at least {ideal_weight} to be healthy!")
elif 18.0 <= BMI <= 24.9:
    print(f"Your BMI is: {BMI}. You are healthy.")
elif 24.9 < BMI <= 29.9:
    print(f"Your BMI is: {BMI}. You are overweight.")
    ideal_weight = 24.9 * (height ** 2)
    ideal_weight = round(ideal_weight, 1)
    print(f"You should reach a weight of at least {ideal_weight} to be healthy!")
elif BMI >= 30:
    print(f"Your BMI is: {BMI}. You are obese.")
    ideal_weight = 24.9 * (height ** 2)
    ideal_weight = round(ideal_weight, 1)
    print(f"You should reach a weight of at least {ideal_weight} to be healthy")
else:
    print("You are obese.")
```
